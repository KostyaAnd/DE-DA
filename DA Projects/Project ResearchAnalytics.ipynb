### Предобработка данных

```python
# Для начала проверим датасет на явные дубликаты

data.duplicated().sum()
```

Явные дубликаты отсутствуют!

```python
# Изучим пропущение значения в столбцах и суммируем

data.isna().sum()
```

В столбцах датафрейма присутствует большое количество пропусков, использование данных из столбцов с пропущенными значениями может пагубно влиять на общий анализ. 
Попробуем заменить значения в колонках с очевидными пропусками:
 - столбец 'balcony' -  заменим на 0 при отстутствиии балкона
 - столбец 'parks_around3000' - заменим на 0 при отстутствиии в трех км.
 - столбец 'ponds_around3000') - заменим на 0 при отстутствиии в трех км.

```python
for column in ['balcony', 'parks_around3000', 'ponds_around3000']:
    data[column] = data[column].fillna(0)

# проверим заполнились ли значения    
data.isna().sum()
```

С остальными значениями сложнее - с целью сохранения медианных значений заполнять пропуски в оставшихся колонках на данном этапе исследования не имеет смысла.

Краткий вывод:

Пропуски в столбцах 'floors_total' - всего этажей в доме и 'locality_name' - название населённого пункта не повлияют на общую выборку.
В столбце 'is_apartment' - апартаменты  для исследования нет значимых значений.
Пропуски в столбцах 'airports_nearest' - расстояние до ближайшего аэропорта в метрах, 'cityCenters_nearest' - расстояние до центра города, 
'parks_around3000' и 'ponds_around3000' имеют идентичные совпадения, возможно из за некорректного заполнения вероятно автоматического.

```python
# Рассмотрим типы данных в каждом столбце:
# зададим формат
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S'
)

# Найдём столбцы, в которых нужно изменить тип данных.
data['days_exposition'] = data['days_exposition'].round()

# Преобразуем тип данных в выбранных столбцах.
for column in ['floors_total', 'balcony', 'parks_around3000', 'ponds_around3000', 'days_exposition']:
    data[column] = data[column].astype('Int64')

    
# Проверим на корректность заполнения

data.info()
display(data.head(10))
```

В столбцах 'floors_total', 'balcony', 'parks_around3000', 'ponds_around3000' и 'days_exposition' (кол-во дней резмещения размещено объявления), необходимо изменить тип данных на целочисленный для удобства расчета.

В столбце 'days_exposition' с целью избежания ошибки с дробью применим метод round()


```python
# Рассмотрим значения в столбце с названиями населённых пунктов.
# Отсортируем и выведем уникальные значения в столбце 'locality_name'.

data['locality_name'].sort_values().unique()
```

```python
# Разобъем столбец 'days_exposition' на категории:
# от 231 дней до 730 дней - долгие продажи;
# от 731 дней до 1095 дней - очень долгие продажи;
# от 1096 дней до 1580 дней - самые долгие продажи.

def cat_days_exposition(days):
    try:
        if 231.0 <= days <= 730.0:
            return 'долгая продажа'
        elif 731.0 <= days <= 1095.0:
            return 'очень долгая продажа'
        elif 1096.0 <= days <= 1580.0:
            return 'самая долгая продажа'
    except:
        pass
```

```python
# Создадим столбец с категориями и посчитаем кол-во значений по каждой:

data['days_exposition_cat_long'] = data['days_exposition'].apply(cat_days_exposition)
data['days_exposition_cat_long'].value_counts()
```
